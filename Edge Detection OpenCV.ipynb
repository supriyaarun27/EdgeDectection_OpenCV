```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

%matplotlib inline 

image=cv2.imread('Desktop/Udacity/edge.jpg')
image_copy = np.copy(image)
#change colors from BGR to RGB 
image_rgb = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
# convert this imsgr to greyscale 
grey = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(grey, cmap='gray')
```

```python
# create a filter for vertical edges 
#sobel filter - change in intensity in the X direction 

sobel_x  = np.array([[ -1, 0 ,1 ],
                     [ -2, 0, 2],
                     [ -1, 0, 1 ]])
f_image_x = cv2.filter2D(grey, -1, sobel_x)
plt.imshow(f_image_x, cmap='gray')
```

```python
#create a filter for horizontal edges 

sobel_y  = np.array([[ -1, -2 ,-1 ],
                     [ 0, 0, 0],
                     [ 1, 2, 1 ]])y
f_image_y = cv2.filter2D(grey, -1, sobel_y)
plt.imshow(f_image_y, cmap='gray')
```

```python
# normal edge dectecion filter - high pass 
hp      = np.array([[ 0, -1 ,0],
                     [ -1, 4, -1],
                     [ 0, -1, 0 ]])
f_image_hp = cv2.filter2D(grey, -1, hp)
plt.imshow(f_image_hp, cmap='gray')
```

```python
# convert an image into pure black and white (0/255)
# gives better clarity on the edges 
retval , binary_image = cv2.threshold(f_image_y, 50, 255 , cv2.THRESH_BINARY)  # everything above 50 is tconvereted to 255, everything below is converted to 0
plt.imshow(binary_image, cmap='gray')
```
